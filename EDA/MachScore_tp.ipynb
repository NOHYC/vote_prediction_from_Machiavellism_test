```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train=pd.read_csv("train.csv")
train=train.drop('index', axis=1)
train.columns
```

```python
train['Extraversion']=round((train.tp01+train.tp06)/2)
train['Agreeableness']=round((train.tp02+train.tp07)/2)
train['Conscientiousness']=round((train.tp03+train.tp08)/2)
train['Emotional Stability']=round((train.tp04+train.tp09)/2)
train['Openness to Experiences']=round((train.tp05+train.tp10)/2)
```

```python
Qaswer = train.iloc[:,::2].iloc[:,:20]
Qaswer
```

```python
Qaswer[["QeA", "QfA", "QkA", "QqA", "QrA", "QaA", "QdA", "QgA", "QiA", "QnA"]] = 6 - Qaswer[["QeA", "QfA", "QkA", "QqA", "QrA", "QaA", "QdA", "QgA", "QiA", "QnA"]]
```

```python
train['MachScore']=round(Qaswer.mean(axis = 1))
train
```

```python
train.groupby(['MachScore', 'Extraversion'])['voted'].value_counts().unstack()
```
