```python
import pandas as pd
train = pd.read_csv(path + 'train_preprocessed.csv')
test = pd.read_csv(path + 'test_preprocessed.csv')
submission = pd.read_csv(path + 'sample_submission.csv')
```

```python
test[['tp02','tp04','tp06','tp08','tp10']]= 6-test[['tp02','tp04','tp06','tp08','tp10']]

test['Extraversion']=(test.tp01+test.tp06)/2
test['Agreeableness']=(test.tp02+test.tp07)/2
test['Conscientiousness']=(test.tp03+test.tp08)/2
test['Emotional Stability']=(test.tp04+test.tp09)/2
test['Openness to Experiences']=(test.tp05+test.tp10)/2
```
